```python
import pandas as pd
import numpy as np
```

```python
data= pd.read_csv('https://gist.githubusercontent.com/curran/a08a1080b88344b0c8a7/raw/639388c2cbc2120a14dcf466e85730eb8be498bb/iris.csv')
```

```python
data.info()
```

```python
data.species.value_counts()
```

```python
features = data.iloc[:,:-1].values
label = data.iloc[:,-1].values
```

```python
# Goal is to create a model with a goal to visualize decision region
# To visualize decision region, we need to use a different package
# mlxtend -- Expects data to be fully numeric (features and label)

finalFeatures = data.iloc[:,[0,2]].values

#Convert the label into discrete values (Reqt for MLXtend)
finalLabels = data.species.replace(['setosa','versicolor','virginica'] , [1,2,3]).values.ravel()
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(finalFeatures,finalLabels)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(finalLabels,model.predict(finalFeatures))
```

```python
#using mlxtend to display decision regions

from mlxtend.plotting import plot_decision_regions
plot_decision_regions(finalFeatures,finalLabels,clf=model,legend=2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model1 = KNeighborsClassifier()
model1.fit(finalFeatures,finalLabels)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(finalLabels,model1.predict(finalFeatures))
```

```python
#using mlxtend to display decision regions

from mlxtend.plotting import plot_decision_regions
plot_decision_regions(finalFeatures,finalLabels,clf=model1,legend=2)
```

```python
from sklearn.svm import SVC
model3 = SVC(kernel="sigmoid")
model3.fit(finalFeatures,finalLabels)
```

```python
#using mlxtend to display decision regions

from mlxtend.plotting import plot_decision_regions
plot_decision_regions(finalFeatures,finalLabels,clf=model3,legend=2)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(finalLabels,model3.predict(finalFeatures))
```

```python
from sklearn.svm import SVC
model4 = SVC(kernel="poly")
model4.fit(finalFeatures,finalLabels)
```

```python
#using mlxtend to display decision regions

from mlxtend.plotting import plot_decision_regions
plot_decision_regions(finalFeatures,finalLabels,clf=model4,legend=2)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(finalLabels,model4.predict(finalFeatures))
```

```python
from sklearn.svm import SVC
model5 = SVC(kernel="rbf")
model5.fit(finalFeatures,finalLabels)
```

```python
#using mlxtend to display decision regions

from mlxtend.plotting import plot_decision_regions
plot_decision_regions(finalFeatures,finalLabels,clf=model5,legend=2)
```
